1. Removing the Nan values and dropping the columns indicating the business data

```python
df = review.dropna()
df = df.reset_index(drop=True)
```

```python
reviews=df.drop(["address","city","state","latitude","longitude","stars_x","review_count","attributes","categories","hours"], axis = 1)
```

```python
reviews.shape
```

```python
reviews=reviews.drop(["index"],axis=1)
```

```python
reviews.head()
```

Grouping the resturants by reviews stars count (more than 100 counts

```python
reviews.groupby('business_id')['stars_y'].count().sort_values(ascending=False)
```

```python
s = reviews.groupby('business_id')['stars_y'].count()
review_users = reviews[reviews.isin(s.index[s >= 100]).values]
```

```python
review_users.groupby('business_id')['stars_y'].count().sort_values(ascending=False)
```

```python
Filter = review_users['business_id'] == 'bZiIIUcpgxh8mpKMDhdqbA'
print("Name: ", review_users[Filter]['name'].unique())
print("The popular resturant in the given YELP datasets according to the reviews is ",review_users[Filter]['name'].unique()[0])
```

```python
# create a user-item matrix 
# now ratings given by each users to the diffeent resturant scope can be visualized as:
```

```python
ratingFromUsers = review_users.pivot_table(values='stars_y', index='user_id', columns='name', fill_value=0)
ratingFromUsers.head()
```

```python
X = ratingFromUsers.transpose()
X.shape
```

<h2>Cosine Similarity on the restaurant data</h2>

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
restaurant_similarity_matrix = cosine_similarity(X)
```

```python
restaurant_similarity_matrix
```

```python
X.head()
```

```python
X = pd.DataFrame(restaurant_similarity_matrix, columns=X.index.values, index=X.index).reset_index()
X
```
