### Cleaning the business data
Taking into account only those business whose review count is no more less than 15 reviews and the business open status is 1.
Also since the column postal code and isopen doennot account so these columns are dropped out.

```python
business=business[(business["is_open"]==1) & (business["review_count"]>=15)]
business = business.drop(columns=["is_open","postal_code"],axis=1)
```

 Now we are only concern with the business which are resturants . Here the datasets contains other variety of business like salon, palour etc so eliminating only those datas having the resturants.

```python
# Resturant data from Business

restaurant = business[business['categories'].str.contains('Restaurant.*')==True].reset_index()
restaurant.head()
```

### Data Visualization
Visualizing the resturants locations based on their latitudes and longitudes

```python
geometry = [Point(xy) for xy in zip(restaurant['longitude'], restaurant['latitude'])]
gdf = GeoDataFrame(restaurant, geometry=geometry)   

#this is a simple map that goes with geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf.plot(ax=world.plot(figsize=(10, 6)), marker='o', color='red', markersize=5);
```

From the above plot most of the resturants are located on the the north america region

```python
# Finding the unique city in which resturant is located 
unique_city=pd.unique(restaurant["city"])
print("The unique city in which the resturants are located are --->",unique_city)
print("-----------")
print("The total numbers of the unique city are ->",len(unique_city))
```

```python
# finding the unique state having the resturant
unique_state=pd.unique(restaurant["state"])
print("The unique city in which the resturants are located are --->",unique_state)
print(len(unique_state))
```

```python
# Total number of resturants in each states
count=restaurant.state.value_counts()
print("The total number of resturants in each states are-->")
print(count)
```

```python
sns.set_theme(style="whitegrid")
tips = sns.load_dataset("tips")
ax = sns.barplot(x=unique_state, y=count)
ax.set_ylabel("frequency", fontsize = 12)
ax.set_xlabel("State", fontsize = 12)
ax.set_title("Total number of restaurant in different state",fontsize=12)
for index, value in enumerate(count.values):
    plt.text(index, value, str(value),ha='center')
```

Visualizing the nuumber of resturants for the given ratings.

```python
fig, ax = plt.subplots(figsize=(15,7))
count_ratings=restaurant.stars.value_counts()
count_ratings=count_ratings.sort_index()
sns.set_theme(style="whitegrid")
ax = sns.barplot(x=count_ratings.index, y=count_ratings)
ax.set_ylabel("frequency", fontsize = 12)
ax.set_xlabel("Ratings", fontsize = 12)
ax.set_title("Total number of restaurant in different for different ratings",fontsize=12)

rects = ax.patches
labels =count_ratings.values
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height + 5, label, ha='center', va='bottom')
```

Here more than 7000 resturants have maximum 4.0 ratings and less than 60 resturants have 1.0 ratings

```python
# Resturants per ratings for different states are:
fig, ax = plt.subplots(figsize=(15,7))
# use unstack()
restaurant.groupby(["state","stars"]).count()['name'].unstack().plot(ax=ax,kind="bar")
ax.set_ylabel("frequency", fontsize = 12)
ax.set_xlabel("States", fontsize = 12)
ax.set_title("Total number of restaurant vs ratings in different state",fontsize=12)
```

 Massachusetts states have the highest number of reviews with greater than 4 ratings

```python
#Visualizations of popular resturants categories
restaurant_cats=' '.join(restaurant['categories'])
# print(restaurant_cats)
cats=pd.DataFrame(restaurant_cats.split(','),columns=['category'])
# print(cats)
unicat=pd.unique(cats["category"])
x=cats.category.value_counts()
print("There are ",len(x)," different types/categories of Businesses in Yelp!")
#prep for chart
x=x.sort_values(ascending=False)
x=x.iloc[0:25]
print("The top categories with their counts are --->\n",x)
#chart
plt.figure(figsize=(16,4))
ax = sns.barplot(x.index, x.values, alpha=0.8)#,color=color[5])
plt.title("The top 25 categories",fontsize=12)
locs, labels = plt.xticks()
plt.setp(labels, rotation=80)
plt.ylabel('Frequencies', fontsize=12)
plt.xlabel('Categories', fontsize=12)

#adding the text labels
rects = ax.patches
labels = x.values
for rect, label in zip(rects, labels):
    height = rect.get_height()
    ax.text(rect.get_x() + rect.get_width()/2, height + 5, label, ha='center', va='bottom')

plt.show()
```

Resturants , Food, Nightlife and other categories have the higher number of resturants.
